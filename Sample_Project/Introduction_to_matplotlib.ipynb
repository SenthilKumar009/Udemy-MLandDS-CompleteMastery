# Matplotlib example workflow

```python
# 0 import matplotlib and it get ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4,5,6]
y = [11,22,33,44,55,66]

#2. Setup Plot
fig, ax = plt.subplots(figsize=(10,10))


#3. Plot the data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

#5. Save and Show
fig.savefig("images/sample_plot.png")
```

# Which one should you use? (pyplot vs matplotlib vs OO method)

* when plotting something quickly, use pyplot
* when plotting something more advanced use OO Method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c = 'target');
```

```python
# OO Method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);

ax.set_xlim(45, 100)
```

```python
## OO method from Scratch
fig, ax = plt.subplots(figsize = (10,6))

## Ploting the data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])

## Customize the plot
ax.set(title = 'Heart disease and Cholesterol Levels',
       xlabel = 'Age',
       ylabel = 'Cholesterol');

## Add a Legend
ax.legend(*scatter.legend_elements(), title='Target');

## Add a Horizantal Line

ax.axhline(over_50['chol'].mean(),
           linestyle = '--')
```

```python
over_50.head()
```

```python
# Subplot of Chol, Age and thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10));


### ax0
scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
# ax1
scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'])

## Customize the Plot
ax0.set(title = 'Heart Diseases and Cholesterol Level',
        xlabel = 'Age',
        ylabel = 'Cholesterol');

ax1.set(title = 'Heart Diseases and Thalach Level',
        xlabel = 'Age',
        ylabel = 'Thalach');

## Add a legend
ax0.legend(*scatter.legend_elements(), title = 'Target');
ax1.legend(*scatter.legend_elements(), title = 'Target');

## Add a horizantal line
ax0.axhline(over_50['chol'].mean(),
           linestyle = '--');
ax1.axhline(over_50['thalach'].mean(),
            linestyle='-');

## Add a title to the figure
fig.suptitle('Heart Diseases Analysis', fontsize = 15, fontweight = 'bold');
```

```python
# Customize the y and x axis limitation

# Subplot of Chol, Age and thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10),
                               sharex = True);


### ax0
scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],
                     cmap = 'winter')
# ax1
scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'],
                      cmap = 'summer')

## Customize the Plot
ax0.set(title = 'Heart Diseases and Cholesterol Level',
        xlabel = 'Age',
        ylabel = 'Cholesterol');


ax1.set(title = 'Heart Diseases and Thalach Level',
        xlabel = 'Age',
        ylabel = 'Thalach');

# change the x-axis
#ax0.set_xlim([50,80])

## Add a legend
ax0.legend(*scatter.legend_elements(), title = 'Target');
ax1.legend(*scatter.legend_elements(), title = 'Target');

## Add a horizantal line
ax0.axhline(over_50['chol'].mean(),
           linestyle = '--');
ax1.axhline(over_50['thalach'].mean(),
            linestyle='-');

## Add a title to the figure
fig.suptitle('Heart Diseases Analysis', fontsize = 15, fontweight = 'bold');
```
